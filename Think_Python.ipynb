## 1.1 Arithmetic Operators

```python
integer_1, integer_2 = 5, 10
float_1, float_2 = 7.0, 3.0

add = integer_1 + integer_2
subtract = integer_2 - integer_1
multiply = integer_1 * integer_2
division = integer_2 / integer_1

print("Addition result & type: ", add, type(add),"\n", 
    "Subtraction result & type: ",subtract, type(subtract), "\n",
    "Multiplication result & type: ", multiply, type(multiply), "\n",
    "Division result & type: ", division, type(division), "\n")  # division of two integers can be float so it is set to float automatically

integer_division_or_floor_division = integer_1 // integer_2  # we get floor value so called floor division

print("Integer division or Floor division result & type: ",integer_division_or_floor_division, type(integer_division_or_floor_division))

exponentiation_int = integer_1 ** integer_2
exponentiation_float = float_1 ** float_2
exponentiation = integer_1 ** float_1


print("Exponentiation result and type: ", exponentiation_int, type(exponentiation_int))
print("Exponentiation result and type: ", exponentiation_float, type(exponentiation_float))
print("exponentiation result and type: ", exponentiation, type(exponentiation))
```

## 1.6 Formal and Natural Languages
Natural language: Languages that we speak
Formal language: Languages that are designed by people for specific applications
    Ex. Programming languages are formal languages
Although formal and natural languages have some features in common there are important differences:

    1. Ambiguity: Natural languages are full of ambiguity, which people deal with by using contextual clues and other information. Formal languages are designed to be nearly or completely unambiguous, which means that any program has exactly one meaning, regardless of context.

    2. Redundancy: In order to make up for ambiguity and reduce misunderstandings, natural languages use redundancy. As a result, they are often verbose. Formal languages are less redundant and more concise.

    3. Literalness: Natural languages are full of idiom and metaphor. Formal languages mean exactly what they say.

```python
# PYTHON OPERATOR PRECEDENCE

print("1️⃣ Parentheses ()  → Grouping")
print("(2 + 3) * 4 =", (2 + 3) * 4)
print()

print("2️⃣ Exponentiation **  → Power (Right-Associative)")
print("2 ** 3 ** 2 =", 2 ** 3 ** 2)
print()

print("3️⃣ Unary Operators (+x, -x, ~x)")
print("+5 =", +5)                  # Unary Plus
print("-5 =", -5)                  # Unary Minus
print("~5 =", ~5)                  # Bitwise NOT
print()

print("4️⃣ Multiplicative (*, /, //, %)")
print("6 * 3 =", 6 * 3)            # Multiplication
print("7 / 2 =", 7 / 2)            # Division
print("7 // 2 =", 7 // 2)          # Floor Division
print("7 % 2 =", 7 % 2)            # Modulus
print()

print("5️⃣ Additive (+, -)")
print("5 + 2 =", 5 + 2)            # Addition
print("5 - 2 =", 5 - 2)            # Subtraction
print()

print("6️⃣ Bitwise Shift (<<, >>)")
print("5 << 1 =", 5 << 1)          # Left Shift
print("8 >> 2 =", 8 >> 2)          # Right Shift
print()

print("7️⃣ Bitwise AND (&)")
print("5 & 3 =", 5 & 3)
print()

print("8️⃣ Bitwise XOR (^)")
print("5 ^ 3 =", 5 ^ 3)
print()

print("9️⃣ Bitwise OR (|)")
print("5 | 3 =", 5 | 3)
print()

print("🔟 Comparisons (==, !=, >, <, >=, <=, in, is)")
print("5 > 3 =", 5 > 3)
print("5 == 3 =", 5 == 3)
print("3 in [1,2,3] =", 3 in [1,2,3])
print()

print("1️⃣1️⃣ Logical NOT (not)")
print("not True =", not True)
print()

print("1️⃣2️⃣ Logical AND (and)")
print("True and False =", True and False)
print()

print("1️⃣3️⃣ Logical OR (or)")
print("True or False =", True or False)
print()

print("1️⃣4️⃣ Assignment (=, +=, etc.)")
x = 5
x += 3
print("x after x += 3:", x)
print()

print("1️⃣5️⃣ Lambda (Anonymous Function)")
square = lambda n: n * n
print("square(4) =", square(4))
```

```python
print(round(42.5), round(43.5), round(44.5), round(45.5))
# Q. If a number ends in 0.5, does Python round up or down?
    # A. Python uses “round half to even” (also called Banker’s Rounding).
    #    So if a number ends in .5, Python rounds to the nearest even number, not always up.
        # If the nearest integer is even → round to that
        # If the nearest integer is odd → round to the other (even) one

# Q. Why python does this
    # A. Because this reduces statistical bias in large datasets.
    # If we always rounded .5 upward, Over many calculations, The results would drift upward,“Round half to even” balances rounding errors.
```

## Debugging
    Three kinds of errors can occur in a program: syntax errors, runtime errors, and semantic errors. It is useful to distinguish between them in order to track them down more quickly.

Syntax error: “Syntax” refers to the structure of a program and the rules about that structure. If there is a syntax error anywhere in your program, Python does not run the program. It displays an error message immediately.

Runtime error: If there are no syntax errors in your program, it can start running. But if something goes wrong, Python displays an error message and stops. This type of error is called a runtime error. It is also called an exception because it indicates that something exceptional has happened.

Semantic error: The third type of error is “semantic”, which means related to meaning. If there is a semantic error in your program, it runs without generating error messages, but it does not do what you intended. Identifying semantic errors can be tricky because it requires you to work backward by looking at the output of the program and trying to figure out what it is doing.
