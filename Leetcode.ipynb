# 268 Missing Number

```python
from typing import List
class Solution:
    def missingNumber(self, nums: List[int]) -> int:
        n = len(nums)
        array = list(range(n+1))
        # print(array)
        for i in array:
            # print(i)
            if i not in nums:
                # print(i)
                return i

sol = Solution()        
a = sol.missingNumber([3,1, 0, 2,5])
```
